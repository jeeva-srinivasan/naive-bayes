```python
import numpy as np
import pandas as pd
import sklearn
```

```python
train_docs=pd.read_csv('movie_review_train.csv')
train_docs.head()
```

```python
train_docs.shape
```

```python
train_docs['class']=train_docs['class'].map({'Pos':1,'Neg':0})
train_docs.head()
```

```python
#convert df into numpy array
train_array=train_docs.values

#X and Y train split
X_train=train_array[:,1]
y_train=train_array[:,0]
y_train=y_train.astype('int')
print(X_train)
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(stop_words='english', min_df=.03, max_df=.8)
```

```python
vec.fit(X_train)
vec.vocabulary_
```

```python
print(len(vec.get_feature_names()))
```

```python
X_transform=vec.transform(X_train)
X_transform
```

```python
test_docs=pd.read_csv('movie_review_test.csv')
test_docs.head()
```

```python
test_docs['class']=test_docs['class'].map({'Pos':1,'Neg':0})
test_docs.head()
```

```python
#convert df into numpy array
test_array=test_docs.values

#X and Y test split
X_test=test_array[:,1]
y_test=test_array[:,0]
y_test=y_test.astype('int')
y_test
```

```python
X_test_transform=vec.transform(X_test)
X_test_transform
```

```python
#bernouli naive bayes
from sklearn.naive_bayes import BernoulliNB 

bnb=BernoulliNB()
```

```python
#fit
bnb.fit(X_transform,y_train)
```

```python
#predict clas
y_pred_class=bnb.predict(X_test_transform)

#predict probability 
y_pred_prob=bnb.predict_proba(X_test_transform)

y_pred_class
```

```python
#accuracy
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
#confusion matrix
metrics.confusion_matrix(y_test,y_pred_class)
```
